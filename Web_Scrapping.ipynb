```python
import requests
from bs4 import BeautifulSoup
import os

base = "http://www.numismaticmarket.com"
klasor_adı = '/content/drive/MyDrive/resimler/e7-20,000,000lira'
os.makedirs(klasor_adı, exist_ok=True)

#for i in range(0, 39):
# Sayfanın URL'si
url = f"http://www.numismaticmarket.com/?cmd=satis&tip=101&sql=p53&sayfa=1&sutun=5&artis=300&seri="

response = requests.get(url)
response.raise_for_status()

# BeautifulSoup kullanarak HTML'i parse et
soup = BeautifulSoup(response.content, 'html.parser')

# Class'ı t1 olan tüm td öğelerini bul
span_elements = soup.find_all('span', attrs={'class': 'YBas11'})

# Bulunan öğelerin içeriğini yazdır
for span in span_elements:
    a_tag = span.find('a')
    if a_tag:
        url = base + a_tag['href']
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        img = soup.find_all('img', attrs={'alt': '[Resmi Tam Görüntüle]'})
        for i in img:
            img_url = base + i['src']
            response = requests.get(img_url)
            response.raise_for_status()
            img_name = img_url.split('/')[-1]  # Resim dosya adını al
            with open(os.path.join(klasor_adı, img_name), 'wb') as f:
               f.write(response.content)
            print(img_url)
```

```python
import requests
from bs4 import BeautifulSoup
import os

klasor_adı = '/content/drive/MyDrive/resimler2'
os.makedirs(klasor_adı, exist_ok=True)

# Sayfanın URL'si
url = "https://tr.wikipedia.org/wiki/T%C3%BCrkiye_Cumhuriyeti_banknotlar%C4%B1"

# Header with User-Agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = requests.get(url, headers=headers)
response.raise_for_status()

# BeautifulSoup kullanarak HTML'i parse et
soup = BeautifulSoup(response.content, 'html.parser')

tr_elements = soup.find_all('tr', attrs={'style': 'text-align:center;'})

for tr in tr_elements:
    a_elements = tr.find_all('a')
    for a in a_elements:
        href = a.get('href', '')
        if href.endswith('.jpg'):
            response_img = requests.get(href, headers=headers)
            response_img.raise_for_status()

            img_soup = BeautifulSoup(response_img.content, 'html.parser')

            img_div = img_soup.find('div', attrs={'class': 'fullImageLink'})
            if img_div:
                img_tag = img_div.find('a')
                if img_tag and img_tag.has_attr('href'):
                    img_link = img_tag['href']
                    full_img_url = f"https:{img_link}"

                    # Resmi indirme ve kaydetme
                    img_response = requests.get(full_img_url, headers=headers)
                    img_response.raise_for_status()

                    img_name = os.path.join(klasor_adı, os.path.basename(full_img_url))
                    with open(img_name, 'wb') as f:
                        f.write(img_response.content)
                    print(f"Resim indirildi: {img_name}")
```
